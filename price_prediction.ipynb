````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
#generating data
np.random.seed(42) #for reproducibility
X = 2*np.random.rand(100, 1) #house sizes (random numbers)
y = 4+3*X+np.random.randn(100,1) #prices with some noise
````

````python
#visualize the data
plt.scatter(X,y,color = "blue")
plt.xlabel("Size (in 1000 sq. ft)")
plt.ylabel("Price(in $1000)")
plt.title("House Prices")
plt.show()
````

````python
#initializing parameters
w = np.random.randn(1)  #random weight
b = np.random .randn(1)  #random bias

#prediction function 
def predict(X,w,b):
    return w*X+b
````

````python
def compute_loss(y_true,y_pred):
    return np.mean((y_true-y_pred)**2)
````

````python
### *Step 5: Implement Gradient Descent*
#Update the parameters 👦w👦 and 👦b👦 iteratively to minimize the loss:  
#```python
# Gradient descent function
def gradient_descent(X, y, w, b, learning_rate, epochs):
    n = len(y)
    for i in range(epochs):
        # Calculate predictions
        y_pred = predict(X, w, b)
        
        # Compute gradients
        dw = -(2/n) * np.sum((y - y_pred) * X)
        db = -(2/n) * np.sum(y - y_pred)
        
        # Update parameters
        w -= learning_rate * dw
        b -= learning_rate * db
        
        # Log progress every 100 epochs
        if i % 100 == 0:
            loss = compute_loss(y, y_pred)
            print(f"Epoch {i}: Loss = {loss}")
    
    return w, b
````

````python
w = np.random.randn(1)
b = np.random.randn(1)
learning_rate = 0.1
epochs = 1000

# Train the model
w, b = gradient_descent(X, y, w, b, learning_rate, epochs)
print(f"Trained weight: {w}, Trained bias: {b}")
````

````python
# Make predictions
X_test = np.array([[1.5], [3.0], [4.5]])  # Test house sizes
y_pred = predict(X_test, w, b)

print("Predicted Prices:", y_pred)

# Plot predictions
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X_test, y_pred, color='red', label='Predictions')
plt.legend()
plt.xlabel("Size (in 1000 sq. ft)")
plt.ylabel("Price (in $1000)")
plt.title("Predictions vs Actual Data")
plt.show()
````
